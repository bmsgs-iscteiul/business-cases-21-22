```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
df_adjClose = pd.read_csv('../data/adj_close.csv')
df_close = pd.read_csv('../data/close.csv')
df_high = pd.read_csv('../data/high.csv')
df_low = pd.read_csv('../data/low.csv')
df_open = pd.read_csv('../data/open.csv')
df_volume = pd.read_csv('../data/volume.csv')
```

```python
#DF LIST TO MERGE
dfs = [df_adjClose, df_close, df_high, df_low, df_open, df_volume]

#GET THE LIST OF THE CRYPTO CURRENCIES
crypto_list = df_adjClose.columns[1:]

#LIST OF MARKET METRICS/OPERATIONS
market_metrics = ['ADJCLOSE','CLOSE','HIGH','LOW','OPEN','VOLUME']
```

```python
#RENAME THE COLUMNS IN THE DF'S WITH THE RESPECTIVE METRIC
i = 0
for df in dfs:
    for crypto in crypto_list:
        df.rename(columns = {crypto : crypto + '_' + market_metrics[i]}, inplace = True)
    i += 1
```

```python
#DATAFRAME WITH ALL OF THE MERGED INFORMATION
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='inner'), dfs)
merged_columns = df_merged.columns
```

```python
#GET THE INDIVIDUAL DATAFRAMES FOR EVERY CRYPTOCURRENCY
crypto_df_list = []

for crypto in crypto_list:
    
    #taking input as a string
    str = "df_" + crypto[:-4]
    
    #Add the df name to the list
    crypto_df_list.append(str)
    
    #Get the columns for the corresponding crypto
    crypto_columns = ['Date']
    for column in merged_columns:
        if crypto[:-4] in column:
            crypto_columns.append(column)
            
    #Assign the df
    locals()[str] = df_merged[crypto_columns]
```
